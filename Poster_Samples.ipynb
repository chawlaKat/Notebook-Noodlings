```
import altair as alt
import bark_to_byte as bb
import scatterable as sc
alt.data_transformers.disable_max_rows()
```

```
cpmp_2015 = bb.arff_to_dataframe('algorithm_runs.arff')
```

```
def join_table(data):
    data.set_index('instance_id', inplace = True),
    data_joined = data.join(data, lsuffix='_x', rsuffix='_y')
    data_joined.reset_index(inplace=True)
    return data_joined
```

```
status_dropdown = sc.make_dropdown(cpmp_2015, 'runstatus', 'status-select')

alt.Chart(cpmp_2015).mark_circle().encode(
    x = 'algorithm',
    y = 'runtime'
).add_selection(
    status_dropdown
).transform_filter(
    status_dropdown
)
```

Show algorithm vs runtime

```
cpmp_2015 = bb.arff_to_dataframe('algorithm_runs.arff')

alt.Chart(cpmp_2015).mark_circle(size=100).encode(
    x = 'algorithm',
    y = 'runtime',
    tooltip = ['instance_id', 'algorithm', 'runtime', 'runstatus']
).interactive()
```

```
cpmp_joined = join_table(cpmp_2015)
cpmp_joined.head()
```

```
sample_chart = sc.create_comparative_scatter(cpmp_joined, 'instance_id', 
                              'algorithm_x', 'algorithm_y',
                              'runtime_x', 'runtime_y',
                              'runstatus_x', 'runstatus_y',
                              ['#ffe200', '#f5ffff', '#6600db', '#11182b'])

sample_chart.save('sample.html')
sample_chart
```
